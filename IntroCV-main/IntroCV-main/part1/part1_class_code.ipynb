# 图像的旋转变换与拉伸变换

## 旋转变换-用原理

```python
print(img.shape)
# 180*0.5/3.14=
theta=1.3
print("angle=%s"%(180*theta/3.14))
cols=726
rows=722
#dst = img.copy()
dst = np.zeros((img.shape),dtype=np.int)

for x1 in range(cols):
    for y1 in range(rows):
         x2 = np.cos(theta)*x1-np.sin(theta)*y1
         y2 = np.sin(theta)*x1+np.cos(theta)*y1
         #dst[y1,x1,:]=0
         #print(dst.shape)
         #print("x2,y2= %s,%s"%(x2,y2))
         if x2>=0 and x2<cols and y2>=0 and y2<rows:
             dst[int(y2),int(x2),:]=img[y1,x1,:]
             
            
plt.imshow(dst)
plt.show()
```

## 旋转变换-调用opencv

```python
#pts1 = np.float32([[50,50],[200,50],[50,200]])
#pts2 = np.float32([[10,100],[200,50],[100,250]])
#M = cv2.getAffineTransform(pts1,pts2)
#print(M)
#180/3.14
theta=0.3
M = np.float32([[np.cos(theta),-np.sin(theta),0],[np.sin(theta),np.cos(theta),0]])
#M = np.float32([[0.1,0.2,100],[0.3,2,100]])
# 变换矩阵，平移，斜切，旋转
# affine
cols=726
rows=722
a=3.5
b=1
#M = np.float64([[a,0,0],[0,b,0]])
print(M)
print(img.shape)
dst = cv2.warpAffine(img,M,(2*cols,3*rows))
plt.imshow(dst)
plt.show()
```

## 拉伸变换/ 透视变换

```python
pts1 = np.float32([[0,0],[100,0],[0,100],[100,100]])
pts2 = np.float32([[0,0],[110,0],[0,100],[100,120]])
M = cv2.getPerspectiveTransform(pts1,pts2)
print(M)
print(img.shape)
# 拉伸变换/透视变换
dst = cv2.warpPerspective(img,M,(1000,1000))
plt.imshow(dst)
plt.show()
```
